# Задание 1

на сайте https://www.ncbi.nlm.nih.gov/datasets/genome/ находим геномы ежа и морского ежа - хочу поискать их сходства

c помощью `datasets download genome accession <id генома вашего любимого организма> --include protein` скачиваем геномы Painted urchin (Lytechinus pictus) и Western European hedgehog (Erinaceus europaeus)



Теперь на сайте https://www.ebi.ac.uk/interpro/entry/pfam/#table ищем любимый белковый домен -- я взяла ATPase family associated with various cellular activities (AAA), скачиваем его HMM и запускаем команды 

`hmmsearch --tblout hedgehog.tbl PF00004.hmm GCF_950295315.1/protein.faa `

`hmmsearch --tblout sea_urchin.tbl PF00004.hmm GCF_037042905.1/protein.faa `

выберем по 2 белка из верхних строк -- у них наименьшие e-value. с помощбю команды grep заберем по 2 белка из каждого генома

`grep -A 1 "XP_063962317.1" GCF_037042905.1/protein.faa > urchin1.fasta`

`grep -A 1 "XP_054758534.2" GCF_037042905.1/protein.faa > urchin2.fasta`

`grep -A 1 "XP_060055232.1" GCF_950295315.1/protein.faa > hedgehog1.fasta`

`grep -A 1 "XP_007530230.1" GCF_950295315.1/protein.faa > hedgehog2.fasta`

Теперь сравним полученные белки множественным выравниванием - я выбрала T-coffee. В целом выглядит нормально, есть вполне схожие участки, но домен видимо не такой консервативный
