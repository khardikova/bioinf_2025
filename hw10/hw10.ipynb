### Бонус

для удобства делаем еще одну папку и кладем в нее копии исходных ридов

затем пишем в командную строку для каждого из двух ридов

`awk '{`

`  if (NR % 4 == 1) { print }`
  
`  else if (NR % 4 == 2) {`
  
`    seq = substr($0, 7, length($0) - 12)`
    
`    print seq`
    
`  }`
  
`  else if (NR % 4 == 3) { print }`
  
`  else if (NR % 4 == 0) {`
  
`    qual = substr($0, 7, length($0) - 12)`
    
`    print qual`
    
`  }`
  
`}' illumina_reads_R2_001.fastq > illumina_reads_R2_001_trimmed.fastq`

и получаем наши trimmed риды. после этого запускаем строку 

`fastqc illumina_reads_R1_001_trimmed.fastq illumina_reads_R2_001_trimmed.fastq`

и получаем QC отчеты о ридах: через Trimmomatics с отрезанием по неподходящему качеству, очевидно, лучше, но второе прочтение в целом выглядит не так плохо
