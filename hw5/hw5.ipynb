### 1

с семинара у нас уже есть транскриптом Luciola cruciata luciferase mRNA, complete cds, сохраненный как lcruciata_luc.fasta, и папка tutorials с сохраненными геномами. Я буду использовать Photinus pyralis (PPYR_OGS1.1.CDS.fa.gz), потому что он мне больше нравится. Применим к нему TransDecoder.LongOrfs и на выходе получим такие файлы





теперь запускаем алгоритм BLAST

`blastp -db ppyr.pep -query lcruciata.pep -outfmt 6 > ppyr_hits.txt`

находим среди них лучшие

`cat acula_hits.txt | head | cut -f2 > acula_hits_best.txt`

и собираем в файл fasta

`seqtk subseq acula/acula.pep acula_hits_best.txt > acula_hits_best.fasta`
